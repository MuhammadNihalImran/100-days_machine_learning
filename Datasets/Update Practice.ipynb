```python
import numpy as np 
import pandas as pd 
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", type(headers_data))
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
from  sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(missing_values=np.nan,strategy='mean')
df[['normalized-losses','bore','stroke','horsepower','peak-rpm']] = imp_mean.fit_transform(df[['normalized-losses','bore','stroke','horsepower','peak-rpm']])
```

```python
imp_most_freq = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df[['num-of-doors']]  = imp_most_freq.fit_transform(df[['num-of-doors']])
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['horsepower'] = df['horsepower'].astype('float')
df['horsepower'].dtype
```

```python
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
group_names = ['low','medium','high']
df['horsepower_binned']  = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower_binned']].head(5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(df["horsepower_binned"])

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
dataset = pd.read_csv('Salary_Dataset.csv')
dataset.head(5)
```

# data train test split 

```python
from sklearn.model_selection import  train_test_split
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
X_train.head(5)
```

# columetansformer and onehotencoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[
    ('encode',OneHotEncoder(),[0])
], remainder='passthrough')
```

```python
X_train = pd.DataFrame(ct.fit_transform(X_train))
X_test  = pd.DataFrame(ct.transform(X_test))
X_train.head(5)
```

```python
columns = ['Australia','Canada','Dubai','USA','Salary','YearsExperience']
X_train.columns = columns
X_test.columns = columns
```
