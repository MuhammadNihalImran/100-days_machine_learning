```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convertRange)
```

# price per square feet

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
```

```python
df['price_per_sqft']
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
location_counts = df['location'].value_counts()
```

```python
location_counts_less_10 = location_counts[location_counts <= 10]
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_counts_less_10 else x)
```

# outlier detect and remove

```python
sns.boxplot(data=df,x='price_per_sqft')
```

```python
def removeOutlier(df, col, show_plot=False):
    percentile25 = df[col].quantile(0.25)
    percentile75 = df[col].quantile(0.75)
    iqr = percentile75 - percentile25

    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr 

    new_df = df[(df[col] < upper_limit) & (df[col] > lower_limit)]

    if show_plot:
        sns.boxplot(data=df, x=col)

    return new_df

# Auto-run until no outliers remain
def remove_outliers_until_clean(df, col, show_plot=False):
    previous_shape = None
    current_df = df.copy()
    iteration = 0

    while previous_shape != current_df.shape:
        previous_shape = current_df.shape
        current_df = removeOutlier(current_df, col, show_plot)
        iteration += 1

    print(f"Outliers removed in {iteration} iterations.")
    return current_df
```

```python
clean_df = remove_outliers_until_clean(df, 'price_per_sqft', show_plot=True)
```

```python
sns.boxplot(data=clean_df,x='price_per_sqft')
```

```python
clean_df1 = remove_outliers_until_clean(clean_df,'bhk',show_plot=True)
```

```python
sns.boxplot(data=clean_df1,x='price_per_sqft')
```

```python
df = clean_df1
```

```python
df.drop(columns=['size','price_per_sqft'],inplace=True)
```

# train and test the data

```python
x = df.drop(columns=['price'])
y = df['price']
x.head(4)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
ct  = ColumnTransformer([
      ("encode",OneHotEncoder(sparse_output=False,handle_unknown='ignore'),['location']) 
  ],remainder="passthrough")
```

```python
scale = StandardScaler()
```

```python
lr =  LinearRegression()
```

```python
pipe = make_pipeline(ct,scale,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
ls = Lasso()
```

```python
pipe = make_pipeline(ct,scale,ls)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
red = Ridge()
```

```python
pipe = make_pipeline(ct,scale,red)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
